1.2 Quantum Bits

A quantum bit, or qubit, for short, is a mathematical object capable of having a state in a linear combination of the states $\lvert 0 \rangle$ or $\lvert 1 \rangle$. This notation is called the Dirac notation, and ensures qubits can be in linear combinations of $\lvert 0 \rangle$ or $\lvert 1 \rangle$, also known as the principle of superposition. As the below equation depicts:

$$\psi = \alpha \lvert 0 \rangle + \beta \lvert 1 \rangle$$

The special states $\lvert 0 \rangle$ and $\lvert1\rangle$ are known as
computational basis states, and form an orthonormal basis for this vector space.
A qubit's state must be normalized to a length of 1, so $|\alpha|^2 + |\beta|^2 = 1$. This implies that the qubit's state us not directly observable, rather, is in a superposition of the two possible quantum states, with corresponding probabilities. As an example, a qubit's state can be represented by $\frac{1}{\sqrt2}\lvert0\rangle +\frac{1}{\sqrt2}\lvert1\rangle$. A key observation here i that the absolute value of the sum of coefficient squares, is indeed equal to 1. Until now, we have depicted qubits as abstract mathematical objects. However, they exist in real-world applications and are represented as the alignment of nuclear spin in a magentic field, or the two possible electron orbits around a stable nucleus, with a ground state $\lvert0\rangle$ and an excited state $\lvert1\rangle$. 

Since $|\alpha|^2 +|\beta|^2 = 1$, we can rewrite the qubit state as:
$$\lvert \psi \rangle = e^{i\gamma}[cos\frac{\theta}{2}\lvert 0 \rangle + e^{i\phi}cos\frac{\theta}{2}\lvert 1 \rangle]$$
However, since the term $e^{i\gamma}$ has no observable effects, our superposition can instead be expressed as:
$$\lvert \psi \rangle = cos\frac{\theta}{2}\lvert 0 \rangle + e^{i\phi}cos\frac{\theta}{2}\lvert 1 \rangle$$
A useful pictorial representation of the superposition is a unit three-dimensional sphere known as a Bloch Sphere, as pictured below. The surface of the bloch sphere represents all possible states of the qubit, which seems to be infinitely many. 



One must note that when a qubit's state is measured, it only yields $\lvert 0 \rangle$ or $\lvert 1 \rangle$. This implies that it collapses from its state of superposition and attains a definite value. This phenomenon cannot be explained, and is rather considered a fundamental postulate of quantum mechanics. 

1.2.1 Multiple Qubits

Let's say we have two qubits. Then, similar to classical bits and their possible combinations, the four possible computational basis states for two qubits are: $$\lvert 00 \rangle, \lvert 01 \rangle, \lvert 10 \rangle, \lvert 11 \rangle$$
Therefore, the two qubits can also exist in a superposition of different states, each with a unique complex coefficient, also known as the amplitude. This superposition can be represented by:
$$\lvert \psi \rangle = \alpha_{00}\lvert 00 \rangle + \alpha_{01}\lvert 01 \rangle + \alpha_{10}\lvert 10 \rangle + \alpha_{11}\lvert 11 \rangle$$
When measured, the probability of state $\lvert x \rangle$ is $|\alpha_x|^2$. Similar to a single-qubit system, the condition that $\Sigma_{x\in(0,1)^2} |\alpha_x|^2 = 1$, known as the normalization condition. 
An important two-qubit state is called the Bell state or EPR pair, described by:
$$\frac{\lvert 00 \rangle + \lvert 11 \rangle}{\sqrt2}$$
A special property of the bell state qubit pair is that the probability of obtaining both 0 and 1 states are each 0.5 and as a result, the measurement of the second qubit always yields the same outcome as the measurement of the first qubit. Hence, these qubit states are correlated stronger than could ever exist in classical systems.
